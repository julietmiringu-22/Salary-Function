```python
#import panda library
import pandas as pd
import os
import zipfile
```

```python
#read the contents of your csv file using the pd.read_csv
data = pd.read_csv("J:/Nexford/R and Python/Module 2/Totalsalary.csv")
```

```python
#display the first few rows and columns of your data using "variable".head()
data.head()
```

```python
#create a dictionary with the employee name is indexed
e_dict = data.set_index('EmployeeName').T.to_dict()
```

```python
# this will create a list of the indexed name together with the other data associated with it
list(e_dict.items())[:2000]
```

```python
# This function retrieves an employees details and when it is not found it displays an error
def get_employee_details(name):
    try:
        return e_dict[name]
    except KeyError:
        return f"Error: Employee '{name}' not found."
```

```python
get_employee_details("Juliet R Martin")
```

```python
# Looks up if employee's name exists 
def safe_employee_lookup(name):
    try:
        if not isinstance(name, str):
            raise ValueError("Name must be a string.")
        
        details = e_dict[name]
        return details
    
    except KeyError:
        return "Employee not found."
    
    except ValueError as ve:
        return f"Input Error: {ve}"
    
    except Exception as e:
        return f"Unexpected Error: {e}"
```

```python
def export_employee_profile(name):
    details = get_employee_details(name)
    
    if isinstance(details, dict):
 # Convert to DataFrame
        emp_df = pd.DataFrame([details])
        
# Create folder on the computer
        folder_name = "Employee Profile"
        os.makedirs(folder_name, exist_ok=True)
        
# Save CSV in the folder
        file_name = f"{folder_name}/{name.replace(' ', '_')}.csv"
        emp_df.to_csv(file_name, index=False)
        
        return file_name
    else:
        return details
```

```python
# creating a zip file in the folder
def zip_employee_folder():
    folder_name = "Employee Profile"
    zip_name = "Employee_Profile.zip"
    
    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for root, dirs, files in os.walk(folder_name):
            for file in files:
                zipf.write(os.path.join(root, file))
    
    return zip_name
```

```python
export_employee_profile("Juliet R Martin")
zip_employee_folder()
```
